## Задание

*Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.*

```python
import matplotlib.pyplot as plt

from PIL import Image
import requests

from io import BytesIO

response = requests.get('https://kartinkin.net/pics/uploads/posts/2022-09/1662780829_31-kartinkin-net-p-koshki-v-turtsii-koti-34.jpg')
img = Image.open(BytesIO(response.content))

plt.imshow(img)
```
