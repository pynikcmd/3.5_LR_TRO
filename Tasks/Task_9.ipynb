Создать ноутбук, в котором выполнить решение вычислительной задачи (например, задачи из области физики, экономики, математики, статистики и т. д.) требующей построения круговой диаграммы, условие которой предварительно необходимо согласовать с
преподавателем.

Условие. В компании A работают мужчины и женщины. Мужчин - 60%, а женщин - 40% в компании. Необходимо проанализировать соотношение мужчин и женщин и представить результаты в виде круговой диаграммы. 

```python
import matplotlib.pyplot as plt

# Данные для построения круговой диаграммы
labels = ['Мужчины', 'Женщины']
sizes = [60, 40]
colors = ['slategrey', 'lightcoral']
explode = (0.1, 0)  # Выделение первого сектора

# Построение круговой диаграммы
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=70)
ax1.axis('equal')

plt.title('Соотношение мужчин и женщин в компании A')

plt.show()
```
