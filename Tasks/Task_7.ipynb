Условие. Даны два города A и B, расстояние между которыми равно 60 км. Из города A одновременно в направлении к городу B выехали два автобуса, первый автобус движется со скоростью 40 км/ч, второй - со скоростью 50 км/ч. Определить расстояние от города B, на котором произойдет встреча автобусов, и время этой встречи. Построим графики зависимости координат автобусов от времени.

```python
import matplotlib.pyplot as plt

# Расстояние между городами А и В
distance = 60

# Скорость первого автобуса
speed_1 = distance / 1.5

# Скорость второго автобуса
speed_2 = distance / 1.2

# Общая скорость движения автобусов
total_speed = speed_1 + speed_2

# Время встречи автобусов
time = distance / total_speed

# Расстояние от точки B
distance_from_B = speed_2 * time

# Вывод результатов
print("Расстояние от точки B: ", distance_from_B, " км")
print("Время встречи автобусов: ", time * 60, " минут")

# Рисуем график
t = [0, time/2, time]
s1 = [0, speed_1*time/2, speed_1*time]
s2 = [distance, distance - speed_2*time/2, 0]

plt.plot(t, s1, 'r-', label='Первый автобус')
plt.plot(t, s2, 'b-', label='Второй автобус')
plt.xlabel('Время, ч')
plt.ylabel('Расстояние, км')
plt.legend()
plt.show()
```
